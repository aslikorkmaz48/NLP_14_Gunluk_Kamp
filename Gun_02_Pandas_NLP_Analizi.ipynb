```python
# 1. Önce gerekli kütüphaneleri yükleyelim (Eğer yüklü değilse)
!pip install zeyrek pandas nltk

import pandas as pd
import re
import nltk
from zeyrek import MorphAnalyzer

# 2. Araçları hazırlayalım (Hata veren paketleri buraya ekledik)
analyzer = MorphAnalyzer()
nltk.download('stopwords')
nltk.download('punkt')      # Genel parçalama için
nltk.download('punkt_tab')  # Senin aldığın hatayı çözen paket
durak_kelimeler = nltk.corpus.stopwords.words('turkish')

# 3. Veri Setini (DataFrame) tekrar oluşturalım
data = {
    'kullanici': ['Aslı', 'Mehmet', 'Can', 'Ece', 'Deniz'],
    'yorum': [
        'Ürün çok kaliteli ve hızlı geldi, teşekkürler.',
        'Kargo çok yavaştı ama ürün beklediğimden iyi çıktı.',
        'Müşteri hizmetleri hiç yardımcı olmadı, iade ediyorum.',
        'Harika bir paketleme, ürün kalitesi muazzam.',
        'Fiyat performans ürünü, herkese tavsiye ederim.'
    ],
    'puan': [5, 4, 1, 5, 4]
}
df = pd.DataFrame(data)

# 4. Temizlik Fonksiyonu
def temizle(metin):
    metin = re.sub(r'[^\w\s]', '', metin.lower()) # Noktalama sil
    kelimeler = metin.split()
    temiz_liste = []
    for k in kelimeler:
        if k not in durak_kelimeler:
            # Zeyrek ile kök bulma
            analiz = analyzer.lemmatize(k)
            kok = analiz[0][1][0]
            temiz_liste.append(kok)
    return " ".join(temiz_liste)

# 5. Tüm tabloyu "yıka" ve yeni sütuna yaz
df['temiz_yorum'] = df['yorum'].apply(temizle)

# Sonucu görelim
print("--- İŞLENMİŞ VERİ TABLOSU ---")
df[['yorum', 'temiz_yorum']]
```

```python
import pandas as pd
import random
import warnings

# 1. Zeyrek'in ekranı dolduran o sarı uyarılarını susturalım
warnings.filterwarnings("ignore")

# 2. 3000 Satırlık veriyi tekrar hızlıca kuralım
yorum_havuzu = [
    "Ürün kalitesi muazzam.", "Kargo çok yavaş.", "Fiyatı pahalı.",
    "Kesinlikle tavsiye ederim.", "İade işlemi çok zor.", "Harika paketleme."
]
df_dev = pd.DataFrame({
    'yorum': [random.choice(yorum_havuzu) for _ in range(3000)],
    'puan': [random.randint(1, 5) for _ in range(3000)]
})

# 3. İŞLEM TAKİBİ: Her 500 satırda bir haber ver diyelim
def temizle_takip(metin, sayac=[0]):
    sayac[0] += 1
    if sayac[0] % 500 == 0:
        print(f"{sayac[0]} satır işlendi...")
    return temizle(metin) # Dünkü fonksiyonu çağırıyoruz

# 4. Uygulayalım
print("İşlem başlıyor...")
df_dev['temiz_yorum'] = df_dev['yorum'].apply(temizle_takip)

# 5. Sonuçları görelim
print("\n--- İŞLEM BİTTİ! ---")
print(df_dev['temiz_yorum'].value_counts().head(10))
```
